# Reading CSV and data cleaning

The quality of data is the most crucial element of any business intelligence strategy. Data cleaning is the process of fixing or removing incorrect, corrupted, incorrectly formatted, duplicate, or incomplete data within a dataset.


```python
!pip install opendatasets --upgrade
```

```python
import opendatasets as od
od.download('https://www.kaggle.com/datasets/himanshupoddar/zomato-bangalore-restaurants')
```

```python

df = pd.read_csv('/content/zomato-bangalore-restaurants/zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

Dropping Duplicates


```python
df.drop_duplicates(inplace = True)
df.shape
```

Cleaning Rate Column

```python
df['rate'].unique()
```

 Removing "NEW" , "-" and "/5" from Rate Column

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

Filling Null Values in Rate Column with Mean

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

Dropping Null Values

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

Listed in(city) and location, both are there, lets keep only one.

```python
df = df.drop(['listed_in(city)'], axis = 1)
df['Cost2plates'].unique()
```

Removing , from Cost2Plates Column

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

Cleaning Rest Type Column

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

Making Rest Types less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

Cleaning Location Column

```python
location = df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

Cleaning Cuisines Column

```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

# Data is clean, lets jump to visualization

## 1. Which locations have the highest count for restaurants? 

Here we are comparing the number of restaurants that are there in different locations of Bangalore. In other words it will be visualization to compare the size of Restaurant market in every area. 

```python
plt.figure(figsize = (16,10));
ax = sns.countplot(df['location']);
plt.xticks(rotation=90);
plt.title("Count Plot of Various Locations");
```

This plot shows us the count for number of restraunts that are there in all the mentioned areas, we can use the above given stats to understand the competition level in the areas and then decide that in which location we need to establish our restraunt. 
Here we can see that Bannerghatta road has the maximum amount of restaurants as compared to other locations. 

## 2. What % of restaurants support online order facility?

Here we are comparing the number of Restaurants that supports online order facility vs the restaurants that do not have this feature. Using this we will be able to judge if there will be advantage by having online order facility or not. 

```python
plt.figure(figsize = (6,6))
sns.countplot(df['online_order'], palette = 'inferno')
plt.title("Visualizing Online Order")
```

From the above figure we can clearly see that there are still a lot of restraunts that do not support online order facility, hence if we will be adding the online order faciity to our restraunt, it will surely help us to establish a stable business. 

## 3. What % of restaurants support online table booking facility?

Here we are comparing the number of Restaurants that supports online table booking facility vs the restaurants that do not have this feature. Using this we will be able to judge if there will be advantage by having online table booking facility or not. 

```python
plt.figure(figsize = (6,6))
sns.countplot(df['book_table'], palette = 'rainbow')
plt.title("Visualizing Book Table")
```

From the above result we can clearly see that there are still a huge number of restraunts that do not support table booking facility so keeping this in mind if we have this feature for our restraunt it will be acting as a big plus point during the long run.

## 4. Which has better rating, online orders or the offline orders? 

Now we are comparing the ratings for restaurants that have online order facility vs the restaurants that don't offer online ordering feature.

From the above mentioned figure we can clearly see that there is a very slight difference in the rating but still the online order facility has the upper hand as it's rating is higher than the offline services

## 5. Which has better rating, restaurants that support online table booking facility or the ones that do not have this option? 

Now we are comparing the ratings for restaurants that have online table booking facility vs the restaurants that don't offer online table booking feature. 


```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'book_table', y = 'rate', data = df)
plt.title("Visualizing Book Table vs Rate")
```

From the above figure we can see that the rating of the restraunts that give table booking facility is way higher than the ones that don't give this facility. So having this feature can add up to the rating of the restrauts in a much better manner. 

## 6. In every location, how many restaurants support online food ordering system as compared to the ones which do not have this option? 


This will be a location wise distribution of restaurants that support online ordering facility, using this we will be able to judge that even if we have this feature then what should be the location that will give us more benefit. 

```python
df1 = df.groupby(['location','online_order'])['name'].count()
df1.to_csv('location_online.csv')
df1 = pd.read_csv('location_online.csv')
df1 = pd.pivot_table(df1, values=None, index=['location'], columns=['online_order'], fill_value=0, aggfunc=np.sum)
df1
```

```python
df1.plot(kind = 'bar', figsize = (10,5))
plt.title("Visualizing Online Order Facility, Location Wise")

```

From this graph we can see that there are a few locations that have less restraunts that support online order facility, so on the basis of this we can choose which area would be best for which type of restraunt, we can choose it to be either online, offline or both on the basis of competiton in the area. 

## 7. According to location, how many restaurants have table booking facility as compared to the ones that don't have this option? 

Here we have visualized location wise distribution of restaurants that have the table booking facility. 


```python
df2 = df.groupby(['location','book_table'])['name'].count()
df2.to_csv('location_booktable.csv')
df2 = pd.read_csv('location_booktable.csv')
df2 = pd.pivot_table(df2, values=None, index=['location'], columns=['book_table'], fill_value=0, aggfunc=np.sum)
df2
```

```python
df2.plot(kind = 'area', figsize = (15,5))
plt.title("Visualizing Book Table Facility, Location Wise")

```

From the above figure we can see that there is a clear differece in almost all the areas for restraunts having the online table booking facility. 
We can see that kammanahalli and cunningham road are having less restraunts that give this facility so we can keep them as out target location. 

## 8. Which type of restaurant have the highest rating?

On the basis of above figure we can conclude that if a restraunt is having options for drinks and have a good ambience or environment for night life have higher rating as compared to the ones that don't have drinks or night timing for the restraunts.

## 10. Plot to see a comparision in the number of people who vote/rate the restaurant after being served? 

This is a location wise distribution of where we will have higher chances of getting ratings whenever a customer is served. 


```python
df4 = df[['location', 'votes']]
df4.drop_duplicates()
df5 = df4.groupby(['location'])['votes'].sum()
df5 = df5.to_frame()
df5 = df5.sort_values('votes', ascending=False)
df5.head()
```

```python
plt.figure(figsize = (15,8))
sns.barplot(df5.index , df5['votes'])
plt.xticks(rotation = 90)
plt.title("No. of Votes, Location Wise")
```

From the above figure we can colclude that there are many areas in which people are very proactive in rating the restraunts so if we can commit to provide a good service everytime then we can choose areas in which people are willing to vote so that the rating of our restraunt can go higher as well. 

## 11. Which are the best selling cuisines? 

```python
df6 = df[['cuisines', 'votes']]
df6.drop_duplicates()
df7 = df6.groupby(['cuisines'])['votes'].sum()
df7 = df7.to_frame()
df7 = df7.sort_values('votes', ascending=False)
df7.head()
```

```python
df7 = df7.iloc[1:, :]
df7.head()
```

```python
plt.figure(figsize = (15,8))
sns.barplot(df7.index , df7['votes'])
plt.xticks(rotation = 90)
plt.title("Visualizing Top Cuisines")
```

From the above given figure we can see that there are certain cuisines that are preffered such as North Indian, Chinese and South Indian. We can choose the type of food that our restaurant has to offer very precisely using this graph.
